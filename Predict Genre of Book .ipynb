```python
seqLengths = [len(sequence.split()) for sequence in df['Name'].to_numpy()]
maxLength = max(seqLengths)
print("Max words in a title is: ", maxLength)
```

```python
from keras.preprocessing.text import Tokenizer
from keras.preprocessing.sequence import pad_sequences
maxWords = 1000
maxLen = maxLength

tok = Tokenizer(num_words = maxWords)
tok.fit_on_texts(df['Name'])
sequences = tok.texts_to_sequences(df['Name'])
matrix = tok.texts_to_matrix(df['Name'], mode = 'binary')
x = pad_sequences(sequences, maxlen = maxLen)
print(len(matrix[0]))
print("Matrix will be used for vanilla neural network: \n", matrix)
print("x will be used for embedding/LSTM \n", x)
```

```python
import numpy as np
genre = df['Genre'].to_numpy()
yDict = {"Fiction": 0, "Non Fiction": 1}
y = np.zeros((len(genre)))
for i in range(len(genre)):
    y[i] = yDict[genre[i]]
print(y)
```

```python
print(x.shape[0]== y.shape[0])
```

```python
#First try vanilla neural network
from keras.layers import Dense, Dropout
from keras.models import Sequential

model = Sequential()
model.add(Dense(64, activation = 'relu',input_shape = (1000,)))


model.add(Dense(1, activation = 'sigmoid'))

model.compile(optimizer = 'rmsprop', loss = 'binary_crossentropy', metrics = ['accuracy'])
model.fit(matrix,y, epochs = 20, validation_split = .2)
```

```python
def textToMatrix(text):
    mat = tok.texts_to_matrix([text], mode = 'binary')
    return mat


def predictWithVanilla(text):
    mat = textToMatrix(text)
    pred = model.predict(mat)
    if(pred[0][0]<0.5):
        print("For the book title of: \"" + text+ "\" =====> predicted fiction")
    else:
        print("For the book title of: \"" + text+ "\" =====> predicted non fiction")
    
predictWithVanilla("lord of the rings")
predictWithVanilla("learn probability theory")
predictWithVanilla("the hobbit")
predictWithVanilla("history of the united states")
predictWithVanilla("working from home")
predictWithVanilla("diet 101")
```

```python
#Next will use word embeddings
from keras.layers import Embedding, Flatten
model2 = Sequential()
model2.add(Embedding(maxWords,8, input_length = maxLen))
model2.add(Flatten())
model2.add(Dense(32, activation = 'relu'))
model2.add(Dense(1, activation = 'sigmoid'))
model2.compile(optimizer = 'rmsprop', loss = 'binary_crossentropy', metrics = ['accuracy'])
model2.fit(x,y, epochs = 50, validation_split = .2)

```

Guess embeddings are not working to well...maybe need more data?

```python
#LSTM
from keras.layers import LSTM
model3 = Sequential()
model3.add(Embedding(maxWords, 32))
model3.add(LSTM(32))
model3.add(Dense(1,activation = 'sigmoid'))
model3.compile(optimizer = 'rmsprop', loss = 'binary_crossentropy', metrics = ['accuracy'])
model3.fit(x,y, epochs = 20, validation_split = .2)
```

Same deal as just using embedding. LSTMs/Embeddings probably require a lot more data as the need to learn more complex relationships.
